```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal_eq(X,y):
  inv = np.linalg.inv(np.dot(X.T ,X))
  theta = np.dot(np.dot(inv, X.T), y)
  probability = np.dot(X, theta)

# normal eq => theta =inverse of (Xtranspose.X).(Xtranspose.y)
```

```python
train_Y = train.loc[:,'label']
train_Y = train_Y.to_numpy()
train_Y
```

```python
train_X = train.iloc[:, 1:21]
train_X = train_X.to_numpy()
train_X
```

```python
normal_eq(y= train_Y, X= train_X) 
```

```python
u = np.ones([np.shape(train_X)[0], 1])
train_X = np.concatenate((u, train_X), axis = 1)
print(np.shape(train_X))
train_Y
```

```python
def grad(X, Y, learningrate, itrn):
  m = Y.size # training set length

  theta = np.zeros( (X.shape[1] , 1) ) # initializing theta

  cost_all = [] # to store cost Function

  # mean normalization

  for i in range( 1, X.shape[1] ):

     mean = np.mean( X[:,i])

     deviate = ( np.amax( X[:,i] ) - np.amin( X[:,i]))

     for j in range(X.shape[0]):

       X[j,i] = ( X[j,i] - mean ) / deviate


  for k in range (Y.size):

    Y[k] =  ( Y[k] - np.mean( Y ) ) / ( np.amax( Y ) - np.amin( Y ))
  

  for z in range(itrn): 

   

      y_probability = np.dot(X, theta) # hypothesis

      cost = (1/(2*m))* np.sum(np.square(y_probability - Y)) # Cost Function

      dif = (1/m)* np.sum(np.dot(X.T, (y_probability - Y) )) # differentiating Cost Function w.r.t Theta

      theta = theta - learningrate*dif # updating theta

      cost_all.append(cost) # updating cost Function list


      # to see if cost is decreasing

      if z % 100 == 0:
        print("Cost is", cost)

  return theta , cost_all
  
```

```python
itrn = 400 #no of interation 
learningrate = 0.0001
theta, cost_all = grad(Y =train_Y ,X = train_X, learningrate = learningrate, itrn = itrn)
```

```python
rng = np.arange(0, itrn)
plt.plot(rng, cost_all)
plt.show()
```
